```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
random_numbers = np.random.randint(1,101,size=100)

plt.plot(random_numbers, 'o')
plt.title("Random Numbers from 1 to 100")
plt.xlabel("Index")
plt.ylabel('Random Number')
plt.show()
```

```python
# Generate random numbers
mean = 50  # Mean 
std_dev = 0.5  # Low standard deviation 
size = 100  # Number of numbers

random_numbers2 = np.random.normal(mean, std_dev, size)

# Verify the standard deviation
calculated_std = np.std(random_numbers2)
print(f"Calculated Standard Deviation: {calculated_std}")

# Plot the numbers
plt.plot(random_numbers2, 'o')
plt.title('Random numbers with low standard devaition')
plt.xlabel('Random Number')
plt.ylabel('Index')
# plt.hist(random_numbers, bins=10, alpha=0.7, color='blue')
# plt.title('Histogram of Random Numbers with Low Std')
# plt.xlabel('Value')
# plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Given series
series = pd.Series([7, 2, 5, 4, 9, 6])

# Step 1: Plot the dataset
plt.figure(figsize=(8, 6))
plt.subplot(3, 1, 1)
plt.plot(series, marker='o', linestyle='-')
plt.title('Dataset')
plt.xlabel('Index')
plt.ylabel('Value')

# Step 2: Calculate and plot the deviations from the mean
mean_value = series.mean()
deviations = series - mean_value
plt.subplot(3, 1, 2)
plt.plot(deviations, marker='o', linestyle='-')
plt.axhline(0, color='black', linewidth=0.5)  # Add horizontal line at y=0
plt.title('Deviations from Mean')
plt.xlabel('Index')
plt.ylabel('Deviation')

# Step 3: Calculate and plot the squared deviations
squared_deviations = deviations ** 2
plt.subplot(3, 1, 3)
plt.plot(squared_deviations, marker='o', linestyle='-')
plt.title('Squared Deviations')
plt.xlabel('Index')
plt.ylabel('Squared Deviation')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Given series
series = pd.Series([7, 2, 5, 4, 9, 6])

# Step 1: Plot the dataset
plt.figure(figsize=(10, 8))

# Plot the dataset
plt.subplot(4, 1, 1)
plt.plot(series, marker='o', linestyle='-')
plt.title('Dataset')
plt.xlabel('Index')
plt.ylabel('Value')

# Step 2: Calculate and plot the deviations from the mean
mean_value = series.mean()
deviations = series - mean_value

plt.subplot(4, 1, 2)
plt.plot(deviations, marker='o', linestyle='-')
plt.axhline(0, color='black', linewidth=0.5)  # Add horizontal line at y=0
plt.title('Deviations from Mean')
plt.xlabel('Index')
plt.ylabel('Deviation')

# Step 3: Calculate and plot the squared deviations
squared_deviations = deviations ** 2

plt.subplot(4, 1, 3)
plt.plot(squared_deviations, marker='o', linestyle='-')
plt.title('Squared Deviations')
plt.xlabel('Index')
plt.ylabel('Squared Deviation')

# Step 4: Calculate standard deviation
standard_deviation = squared_deviations.mean() ** 0.5

# Plot standard deviation
plt.subplot(4, 1, 4)
plt.axhline(standard_deviation, color='red', linestyle='--', label='Standard Deviation')
plt.title('Standard Deviation')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
